```python
import torch

import numpy as np
```

```python
# path = "DVHL__ppo__0512154751__3407__inference"
path = "DVHL__ppo__0512163107__3407__inference"

n_neighbours = []
MN_ratios = []
FP_ratios = []
for i in range(32):
    n = torch.load("./runs/{}/n_neighbours_{}.pt".format(path, i))
    n_neighbours.append(n)

    mn = torch.load("./runs/{}/MN_ratio_{}.pt".format(path, i))
    MN_ratios.append(mn)

    fp = torch.load("./runs/{}/FP_ratio_{}.pt".format(path, i))
    FP_ratios.append(fp)
```

```python
for i in range(32):
    print(np.mean(n_neighbours[i]), np.std(n_neighbours[i]), np.min(n_neighbours[i]), np.max(n_neighbours[i]))
```

```python
for i in range(32):
    print(np.mean(MN_ratios[i]), np.std(MN_ratios[i]), np.min(MN_ratios[i]), np.max(MN_ratios[i]))
```

```python
for i in range(32):
    print(np.mean(FP_ratios[i]), np.std(FP_ratios[i]), np.min(FP_ratios[i]), np.max(FP_ratios[i]))
```
